# Оператори в Python   
 
= Присвоює значення правого операнда лівому.  
+= Додає значення правого операнда до лівого і присвоює цю суму лівому операнду.  
-= Віднімає значення правого операнда від лівого та присвоює результат лівому операнду.  
*= Множить правий операнд на лівий і присвоює результат лівому операнду.  
/= Ділить лівий операнд на правий і присвоює результат лівому операнду.  
%= Ділить по модулю операнди і присвоює результат лівому.  
**= Підносить лівий операнд до степені правого та присвоює результат лівому операнду.  
//= Знаходить цілу частину від ділення лівого операнда на правий і присвоює результат лівому операнду.  

````python
a = 10
b = 2
a //= b # a = a // b
a
````

````python
from decimal import Decimal, ROUND_HALF_DOWN

d = Decimal('1.1364') # звертаємося без decimal., т.к. імпортували Decimal
print(d.quantize(Decimal("1.00"))) # округлення до 2 знаків після коми
````

````python
# ROUND_HALF_UP: округляє число в сторону збільшення, якщо після нього 5 або більше
print(d.quantize(Decimal("1.00"), ROUND_HALF_DOWN))
````

````python
# ROUND_CEILING - округлення в напрямку нескінченності (Infinity);
# ROUND_FLOOR - округляє в напрямку мінус нескінченності (- Infinity);
# ROUND_DOWN - округлення в напрямку нуля;
# ROUND_HALF_EVEN - округлення до найближчого парного числа.
# Число 4.9 округліться не до 5, а до 4 (бо 5 - не парне);
# ROUND_HALF_DOWN - округлення до найближчого нуля;
# ROUND_UP - округлення від нуля;
# ROUND_05UP — заокруглення від нуля (якщо остання цифра після заокруглення до нуля була б 0 або 5, інакше до нуля).
````
