````python
n = 6
while n > 0:
    # check even and odd
    if n % 2 == 0:
        print(n, 'is a even number')
    else:
        print(n, 'is a odd number')
    # decrease number by 1 in each iteration
    n = n - 1
````

````python
i = 1
# outer while loop
# 4 rows in pattern
while i < 5:
    j = 0
    # nested while loop
    while j < i:
        print('*', end=' ')
        j = j + 1
    # end of nested while loop
    # new line after each row
    print()
    i = i + 1
````

## 

**`range(start, stop, step_size)`**

* start -   (')
* stop -    (')
* step_size -  (')

````python
range(3)  # [0, 1, 2]
print(range(3))
print(type(range(3)))
print(list(range(3)))
print(list(range(1, 4)))
print(list(range(1, 4, 2)))  # start, stop[, step]
````

````python
#     20
sum = 0
for i in range(21):
    # print(i, end=' ')
    sum = sum + i
print(sum)
````

````python
#     10  20
sum = 0
for i in range(10, 21):
    # print(i, end=' ')
    sum = sum + i
print(sum)
````

````python
#      10  20
sum = 0
for i in range(10, 21, 2):
    sum = sum + i
print(sum)
````
