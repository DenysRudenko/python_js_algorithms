# Умовна конструкція if-else

```
if умова_істина:      
    оператори                   # які оператори відносяться до if визначає відступ                  
else:                  
    оператори                   # які оператори відносяться до else визначає відступ          
```



````python
# python program to illustrate If else statement
#!/usr/bin/python

i = 20

if i < 15:
    print("i is smaller than 15")
    print("i'm in if Block")
else:
    print("i is greater than 15")
    print("i'm in else Block")
print("i'm not in if and not in else Block")
````

# Вкладені if-else

```
if умова_істина:
    оператори                 # необов'язкова частина
    if умова_істина:      
        оператори                                     
    else:                     # необов'язкова частина
        оператори          
    оператори                 # необов'язкова частина
else:                  
    оператори         
```



````python
i = 10
if i == 10:
   
    #  First if statement
    if i < 15:
        print("i is smaller than 15")
         
    # Nested - if statement
    # Will only be executed if statement above
    # it is true
    if i < 12:
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
````

````python
year = 1900
if year % 400 == 0:
    print('Високосний')
else:
    if year % 100 == 0:
        print('Невисокосний')
    else:
        if year % 4 == 0:
            print('Високосний')
        else:
            print('Невисокосний')
````

### Скорочений запис інструкції if-else

```
if X:
    A = Y
else:
    A = Z
    
Рівносильно інструкції: A = Y if X else Z
```

````python
# У змінну a передати 1, якщо num >= 0, а якщо num < 0, то підставити 0

num = int(input())
a = 0 if num < 0 else 1

print('num:', num,', a:',a)
````

# Цикл while

````python
# Порахувати суму чисел від 1 до 3

sum_total = 0
i = 1

#sum_total = 0 + 1 + 2 +3
while i <= 3:
    sum_total += i
    print(f'for step {i} sum_total = {sum_total}')
    i += 1
          
print(sum_total)  
````

## Умова всередині циклу

````python
n = 6
while n > 0:
    # check even and odd
    if n % 2 == 0:
        print(n, 'is a even number')
    else:
        print(n, 'is a odd number')
    # decrease number by 1 in each iteration
    n = n - 1
````

## Вкладені цикли

````python
i = 1
# outer while loop
# 4 rows in pattern
while i < 5:
    j = 0
    # nested while loop
    while j < i:
        print('*', end=' ')
        j = j + 1
    # end of nested while loop
    # new line after each row
    print()
    i = i + 1
````

## Діапазон

**`range(start, stop, step_size)`**

* start - початок відліку (необов'язково)
* stop - верхню границя відліку (обов'язково)
* step_size - крок (необов'язково)

````python
range(3)  # [0, 1, 2]
print(range(3))
print(type(range(3)))
print(list(range(3)))
print(list(range(1, 4)))
print(list(range(1, 4, 2)))  # start, stop[, step]
````

````python
# Порахувати суму чисел до 20
sum = 0
for i in range(21):
    # print(i, end=' ')
    sum = sum + i
print(sum)
````

````python
# Порахувати суму чисел від 10 до 20
sum = 0
for i in range(10, 21):
    # print(i, end=' ')
    sum = sum + i
print(sum)
````

````python
# Порахувати суму парних чисел від 10 до 20
sum = 0
for i in range(10, 21, 2):
    sum = sum + i
print(sum)
````

````python
# Обернений порядок
num = 3
for num in range(num, -1, -1):
    print(num)
````

# Оператор continue
```
continue - примусовий перехід на наступне коло виконання циклу
```


````python
# continue
for i in range(10):    
    if i % 2 == 1:
        continue
    print(i, end=' ')
````

# Оператор break
```
break - примусове закінчення виконання циклу
```


````python
# break
for i in range(10):    
    if i % 2 == 1:
        break
    print(i, end=' ')
````

# Оператор else в циклах

```
while умова_істина:
    оператори
else:
    оператори
```

```
for лічильник in діапазон:
    оператори
else:
    оператори
```

оператори в блоці else виконується за умови, що вихід з циклу відбувся без використання break

````python
for i in range(1, 6):
    print(i)
else:
    print("Done")
````

````python
count = -100
for i in range(1, 6):
    count = count + 1
    if count > 2:
        break
    else:
        print(i)
else:
    print("Done")
````
