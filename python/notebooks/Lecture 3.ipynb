# Порівняння рядків

Слова порівнюються за словниковим правилом, який визначається порядком символів.
```
ord(ch)	Порядковий номер символа ch
chr(n)  Символ з порядковим номером n
```

* Великі латинські літери менші за маленькі.
* Латинські літери одного регістру впорядковані в алфавітному порідку.
* Цифри-символи менші за літери.
* Цифри-символи впорядковані у порядку зростання.

````python
print(ord("0"))    # 48
print(ord("A"))    # 65
print(ord("a"))    # 97
print(chr(48))     # A
print(chr(65))     # A
print(chr(97))     # a
````

# Форматування чисел  
```
Тип	Значення
c   Відповідний символ Unicode
b   Двійковий формат
o   Вісімковий формат
x   Шістнадцятковий формат (у нижньому регістрі)
X   Шістнадцятковий формат (у верхньому регістрі)
n   Те саме, що й d, але використовує місцеве налаштування для поділу числа
e   Експонентний запис (e у нижньому регістрі)
E   Експонентний запис (E у верхньому регістрі)
f   Відображати фіксовану кількість символів (за замовченням 6)
F   Те саме, що і f, тільки відображає inf як INF, а nan як NAN
g   Загальний формат. Округлює число до p цифр (точність за замовчуванням: 6)
G   Те саме, що і g. Але перемикається до E, якщо число дуже велике
%   Відсотки. Ділить на 100 і додає % наприкінці
```

````python
acc = 0.890152
prec = 0.677787
rec = 0.881254
print('Accuracy is {0:%}, Precision is {1:.1%}, Recall is {2:.3%}'.format(acc, prec, rec))
````

````python
for x in range(1, 11):
    print('{0:2d} {1:3d} {2:4d}'.format(x, x**2, x**3))
````

````python
format_string = '{:.2f} {:.8f} {:.4f}'
for x in range(1, 11):
    y = x/10
    print(format_string.format(y, y**2, y**3))
````

## Вирівнювання
```
'<' Символи-заповнювачі будуть праворуч (вирівнювання об'єкта по лівому краю) (за замовчуванням).
'>' вирівнювання об'єкта з правого краю.
'^' Вирівнювання по центру.
```


````python
'|{:^20}|'.format('aligned')
````

````python
'|{:>20}|'.format('aligned')
````

````python
'|{:<20}|'.format('aligned')
````

````python
'|{:-^20.1f}|'.format(2.345)
````

````python
f'|{2.345:-^20.1f}|'
````
